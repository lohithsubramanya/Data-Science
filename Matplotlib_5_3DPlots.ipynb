```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

# meshgrid - makes 2 dimensional matrix using 2 1-D matrix
X,Y = np.meshgrid(x,y)
```

```python
X.shape
```

```python
Y.shape
```

```python
X.ndim
```

```python
Y.ndim
```

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x) # To show some pattern in the data
```

```python
Z = f(X,Y)
```

```python
Z.shape
```

```python
Z.ndim
```

```python
plt.contour(X,Y,Z,color='black')
```

```python
plt.contour(X,Y,Z,color='black')
plt.colorbar()
```

```python
plt.contour(X,Y,Z,colors='black')  

## Notice the difference because of using the argument 'colors' instead of 'color' 

plt.colorbar()
```

```python
plt.contour(X,Y,Z, 20, cmap='RdGy')
plt.colorbar()
```
