```python
fig = plt.figure()  ## creates an oject for figure
ax = plt.axes() ### creates axes with default limits as 1 on each axes
x = np.linspace(0,10,1000)  ### generate points 
ax.plot(x,np.sin(x))  ## plot sin curve
ax.plot(x,np.cos(x)) ## plots cos curve
```

# Line Color and Styles

```python
plt.plot(x,np.sin(x),color='red')  ## To specify color
```

```python
plt.plot(x, np.sin(x), color='#91487B')   

### Color ID taken from "http://htmlcolorcodes.com/"
```

```python
plt.plot(x, np.sin(x), color='#91487B', linestyle='-.')
```

```python
plt.plot(x, np.sin(x), color='#91487B', linestyle='--')
```

```python
plt.plot(x, np.sin(x),'-.r')  

## Combining linstyle with color red
```

# Axes Limits

```python
plt.plot(x, np.sin(x))
plt.xlim(-5,10)
plt.ylim(0,20)
```

```python
plt.plot(x, np.cos(x))
plt.xlim(-5,10)
plt.ylim(0,20)
```

```python
plt.plot(x, np.tan(x))
plt.xlim(-5,10)
plt.ylim(0,20)
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')   ## to equally space the curve in the graph
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.title('My matplot graph')  ## To insert Title for the graph
plt.xlabel('x')   ## X axis label
plt.ylabel('sin(x)')  ## Y axis label
```

```python
plt.plot( x, np.sin(x), label='sin(x)')
plt.plot( x, np.cos(x), label='cos(x)')
plt.axis('equal')

# plotting the legend
plt.legend()
```
