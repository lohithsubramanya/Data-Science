# The right way [Because the above method of getting score = 1 is not feasible]

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
X = iris.data
y = iris.target
```

```python
X1,X2,y1,y2 = train_test_split(X,y,random_state=0,train_size=0.75)
```

```python
X1.shape
```

```python
X2.shape
```

```python
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X1,y1)
```

```python
y2_predict = model.predict(X2)
```

```python
from sklearn.metrics import accuracy_score #Finding accuracy of classification algo

y_model = model.predict(X)
accuracy_score(y, y_model)
accuracy_score(y2,y2_predict)
```

```python
model.predict([[ 1,  1,  1, 1]])
```

# Validation Curves in Scikit Learn 

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
def Polynomialregression(degree = 2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),LinearRegression(**kwargs))
```

```python
import numpy as np
```

```python
### Some random function to generate X and y 

def make_data(N, err=1.0, rseed=1):
    # randomly sample the data
    rng = np.random.RandomState(rseed)
    X = rng.rand(N, 1) ** 2
    y = 10 - 1. / (X.ravel() + 0.1)
    if err > 0:
        y += err * rng.randn(N)
    return X, y

X, y = make_data(40)
```

```python
X
```

```python
y
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # plot formatting
```

```python
X_test = np.linspace(-0.1,1.1,500)[:,None] ## [:,None] is same as [:, np.newaxis]
```

```python
X_test
```

```python
plt.scatter(X.ravel(),y)

axis = plt.axis()
for degree in [1,3,5]:
    y_predict = Polynomialregression(degree).fit(X,y).predict(X_test)
    plt.plot(X_test.ravel(),y_predict,label='degree{0}'.format(degree))
plt.legend()
```

```python
from sklearn.learning_curve import validation_curve
degree = np.arange(0,40)

train_score,val_score = validation_curve(Polynomialregression(),X,y,'polynomialfeatures__degree',degree,cv=7)
```

```python
train_score  ## Notice the 7 columns due to cv = 7
```

```python
val_score
```

```python
plt.plot(degree, np.median(train_score, 1), color='blue', label='training score')
plt.plot(degree, np.median(val_score, 1), color='red', label='validation score')
plt.ylim(0, 1)
plt.xlabel('degree')
plt.ylabel('score');
plt.legend(loc='best')
```
