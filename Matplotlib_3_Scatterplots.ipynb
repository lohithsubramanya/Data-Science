```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

# alpha - transperancy of colors
# cmap - color coding option
# colors - choosing colors from there

plt.scatter(x,y, c=colors, s=sizes, alpha=0.3, cmap='plasma')
plt.colorbar()  ## To guage the intensity of each color
```
