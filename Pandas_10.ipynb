# Pandas Time Series : Indexing by Time

#### The purpose here is to use datetime as index column and not a regular column
#### Before using datetime as index, convert date column as index

```python
index = pd.DatetimeIndex(['2017-07-04','2017-08-4','2017-09-04','2017-10-04'])
```

```python
data = pd.Series([0,1,2,3], index=index)
```

```python
data.head()
```

```python
data['2017-07-04 ':'2017-09-04']
```

```python
index = pd.DatetimeIndex(['2017-07-04','2018-08-4','2017-09-04','2018-10-04'])
data = pd.Series([0,1,2,3], index=index)
```

```python
data
```

```python
data['2017']
```

```python
data['2018']
```

```python
pd.date_range('2017-02-26', '2017-03-02')
```

```python
pd.date_range('2017',periods=10,freq='A')  ## A - Annualy , periods = 10 means 10 iterations
```

```python
pd.period_range('2018-05',periods=8, freq='M')
```

```python
pd.timedelta_range(0, periods=10, freq='H')
```

```python
date = pd.to_datetime("5th of July 2019")
```
